```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up plotting
%matplotlib inline
sns.set(style='whitegrid')
```

```python
# Set your data directory
DATA_DIR = 'train'

# List process and metrology files
process_files = sorted([f for f in os.listdir(DATA_DIR) if f.startswith('incoming_run_data')])
metrology_files = sorted([f for f in os.listdir(DATA_DIR) if f.startswith('metrology_data')])

print("Process files:", process_files[:5], "...")
print("Metrology files:", metrology_files[:5], "...")
```

```python
# Load a sample process and metrology file
process_sample = pd.read_parquet(os.path.join(DATA_DIR, process_files[0]))
metrology_sample = pd.read_parquet(os.path.join(DATA_DIR, metrology_files[0]))

display(process_sample.head())
display(metrology_sample.head())
display(process_sample.info())
display(metrology_sample.info())
```

```python
print("Process sample shape:", process_sample.shape)
print("Process columns:", process_sample.columns.tolist())
print(process_sample.info())

print("\nMetrology sample shape:", metrology_sample.shape)
print("Metrology columns:", metrology_sample.columns.tolist())
print(metrology_sample.info())
```

```python
# Check for missing data
print(process_sample.isnull().sum())
print(metrology_sample.isnull().sum())
```

```python
# Unique values in categorical columns
for col in ['Tool Id', 'Recipe', 'Process Step', 'Sensor Name']:
    if col in process_sample.columns:
        print(f'Unique {col}:', process_sample[col].unique())
```

```python
# Plot one sensor's value across time for the first run
sensor = process_sample['Sensor Name'].unique()[0]
df_sensor = process_sample[process_sample['Sensor Name'] == sensor]

plt.figure(figsize=(10,4))
plt.plot(df_sensor['Time Stamp'], df_sensor['Sensor Value'])
plt.xlabel('Time Stamp')
plt.ylabel('Sensor Value')
plt.title(f'Sensor: {sensor} in {process_files[0]}')
plt.show()
```

```python
# Distribution
plt.figure(figsize=(6,4))
metrology_sample['Measurement'].hist(bins=20)
plt.xlabel('Measurement')
plt.title('Measurement Value Distribution')
plt.show()

# Spatial plot (X vs Y, colored by Measurement)
plt.figure(figsize=(6,6))
sc = plt.scatter(metrology_sample['X'], metrology_sample['Y'], c=metrology_sample['Measurement'], cmap='viridis')
plt.colorbar(sc, label='Measurement')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Spatial Distribution of Measurements')
plt.show()
```

```python
# Combine all process and metrology files
all_process = pd.concat([pd.read_parquet(os.path.join(DATA_DIR, f)) for f in process_files], ignore_index=True)
all_metrology = pd.concat([pd.read_parquet(os.path.join(DATA_DIR, f)) for f in metrology_files], ignore_index=True)

print("Combined process shape:", all_process.shape)
print("Combined metrology shape:", all_metrology.shape)
```

```python
# Aggregate sensor stats per Run, Sensor
agg = process_sample.groupby(['Run ID', 'Sensor Name'])['Sensor Value'].agg(['mean', 'std', 'min', 'max', 'median'])
display(agg.head())
```

```python
# Pivot so each (Run ID) is a row, columns are sensor mean values
sensor_means = process_sample.pivot_table(
    index='Run ID',
    columns='Sensor Name',
    values='Sensor Value',
    aggfunc='mean'
)
display(sensor_means.head())
```

```python
# Correlation Matrix
# Only keep numeric columns
corr_matrix = sensor_means.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Sensor Means')
plt.show()
```

```python
# Correlation with Target (Metrology)
# First, aggregate metrology by Run (mean measurement per run)
metrology_run = metrology_sample.groupby('Run ID')['Measurement'].mean()

# Join with sensor features
eda_df = sensor_means.join(metrology_run, on='Run ID')

# Correlation with Measurement
correlations = eda_df.corr()['Measurement'].sort_values(ascending=False)
print("Correlation with Measurement:\n", correlations)
```

```python
# Temporal Trends and Lag Features
# Example: mean sensor value over time for one Tool
tool = process_sample['Tool ID'].unique()[0]
tool_df = process_sample[process_sample['Tool ID'] == tool]

pivot_time = tool_df.pivot_table(index='Run Start Time', columns='Sensor Name', values='Sensor Value', aggfunc='mean')
pivot_time.plot(figsize=(12,6))
plt.title(f"Sensor Mean Values Over Time for Tool {tool}")
plt.ylabel('Sensor Value')
plt.show()
```

```python
# Feature Distributions and Boxplots
sns.boxplot(x='Process Step', y='Sensor Value', data=process_sample)
plt.title('Sensor Value by Process Step')
plt.show()

sns.boxplot(x='Sensor Name', y='Sensor Value', data=process_sample)
plt.title('Distribution of Sensor Value by Sensor')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Drop rows with missing target or features
eda_clean = eda_df.dropna(subset=['Measurement'])

X = eda_clean.drop('Measurement', axis=1)
y = eda_clean['Measurement']

model = RandomForestRegressor(n_estimators=50, random_state=42)
model.fit(X, y)

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('Feature Importances (Random Forest)')
plt.show()
```

# Sample Training

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# 1. Load your data (replace with your file paths)
process_sample = pd.read_parquet('train/incoming_run_data_1.parquet')
metrology_sample = pd.read_parquet('train/metrology_data1.parquet')
```

```python
# 2. Aggregate process data per Run ID for each sensor (mean, std, min, max)
features = process_sample.pivot_table(
    index='Run ID',
    columns='Sensor Name',
    values='Sensor Value',
    aggfunc=['mean', 'std', 'min', 'max']
)
# Flatten multi-level columns
features.columns = ['_'.join(col).strip() for col in features.columns.values]
features.reset_index(inplace=True)
```

```python
# 3. Join metrology data with process features on Run ID (point-wise join)
data = pd.merge(metrology_sample, features, on='Run ID', how='left')
```

```python
# 4. Add categorical columns from process_sample to data (if needed for one-hot encoding)
# Find categorical columns
categorical_cols = []
for col in ['Process Step', 'Tool ID', 'Recipe', 'Step ID']:
    if col in process_sample.columns:
        categorical_cols.append(col)

# Merge the categorical columns into 'data' for each Run ID
if categorical_cols:
    # Drop duplicates to avoid merge issues
    cat_df = process_sample[['Run ID'] + categorical_cols].drop_duplicates(subset=['Run ID'])
    data = pd.merge(data, cat_df, on='Run ID', how='left')
```

```python
# 5. One-hot encode categorical columns
if categorical_cols:
    X_categoricals = pd.get_dummies(data[categorical_cols], drop_first=True)
else:
    X_categoricals = pd.DataFrame(index=data.index)  # empty DataFrame if no categoricals
```

```python
# 6. Prepare numeric features (exclude target and non-features)
feature_cols = [col for col in data.columns if col not in ['Run ID', 'X', 'Y', 'Measurement'] + categorical_cols]
X_numeric = data[feature_cols].select_dtypes(include=[np.number])
```

```python
# 7. Combine numeric and categorical features
X_full = pd.concat([X_numeric, X_categoricals], axis=1)
```

```python
# 8. Prepare target
y = data['Measurement']
```

```python
# 9. Fill missing values
X_full = X_full.fillna(X_full.mean())
```

```python
# 10. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size=0.2, random_state=42)
```

```python
# 11. Train the model
model = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)
```

```python
# 12. Predict and evaluate
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Test RMSE: {rmse:.4f}")
```

```python
# 13. (Optional) Feature importance
import matplotlib.pyplot as plt

feat_importances = pd.Series(model.feature_importances_, index=X_full.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importances')
plt.show()
```
